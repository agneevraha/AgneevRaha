# Matplotlib

```python
# import libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Load `mtcars` Dataset

```python
df = pd.read_csv('mtcars.csv')
```

## Check the Data

```python
df.head()
```

## Univariate Plots Using Matplotlib

```python
# boxplot
plt.boxplot(df['mpg'])
plt.title('Boxplot of Miles per Gallon')
plt.ylabel('mpg');
```

```python
# histogram
plt.hist(df['disp'])
plt.title('Histogram of Engine Displacement')
plt.xlabel('Engine Displacement in cc')
plt.ylabel('Count');
```

```python
# Lineplot
plt.plot(df['hp']);
```

## Bivariate Plots using Matplotlib

```python
# scatterplot
plt.scatter(df['wt'],df['mpg'])
plt.title('weight vs mpg')
plt.xlabel('Weight of the car')
plt.ylabel('Mileage of the car');
```

```python
plt.scatter(df['disp'],df['hp']);
```

```python
# practice scatter plots between other pairs of numeric features in the data
```

## Subplots Using Matplotlib

```python
# plt.subplots
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(10,6)) # plt.subplots(1,3)
ax[0].boxplot(df['mpg'])
ax[1].hist(df['mpg'])
plt.suptitle('Miles per Gallon');
```

```python
# plt.subplot
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.boxplot(df['mpg'])
plt.subplot(122)
plt.hist(df['mpg'])
plt.suptitle('Miles per Gallon');
```

## Univariate Plots using Seaborn

```python
# boxplot
sns.boxplot(data=tips,x='total_bill');
```

```python
# histogram
sns.histplot(data=tips, x='tip');
```

```python
# density plot
sns.kdeplot(data=tips, x='tip');
```

```python
# countplot
sns.countplot(data=tips, x='day',order=['Sat','Sun','Thur','Fri']);
```

```python
sns.countplot(data=tips,x='day', order=tips['day'].value_counts().index);
```

```python
# lineplot
sns.lineplot(tips['total_bill']);
```

## Bivariate Plots using Seaborn

```python
# boxplot
sns.boxplot(data=tips, x='day',y='total_bill');
```

```python
sns.__version__
```

```python
# barplot
sns.barplot(data=tips, x='day',y='total_bill',ci=False)
plt.ylabel('Avg Bill Amount($)');
```

```python
# pointplot
sns.pointplot(data=tips, x='day', y='total_bill',ci=False)
plt.ylabel('Avg Bill Amount');
```
