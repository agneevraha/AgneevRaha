# TSNE

## Importing the libraries

```python
# import the necessary packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")

from sklearn.manifold import TSNE
```

## Loading the Dataset

```python
# load the dataset TNSE_data.csv
df = pd.read_csv('TSNE_data.csv')
df.head()
```

```python
df.shape
```

## Segregating the target variable from the rest of the dataset

```python
# seggregate the target variable and the rest of the features
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

## Implementing TSNE

```python
# create a TSNE object with 2 components and transform the data
tsne = TSNE(n_components=2, random_state=0)
X_2d = tsne.fit_transform(X)
```

## Transformed Reduced Dimension Data

```python
X_2d
```

```python
X_2d.shape
```

```python
# create a dataframe with the 2 components and add the target variable
df_2d = pd.DataFrame(X_2d, columns=['t1', 't2'])
df_2d['diagnosis'] = y
df_2d.head()
```

## Visualizing the results of TSNE in terms of the target variable classes

```python
import seaborn as sns
sns.scatterplot(x='t1', y='t2', hue='diagnosis', data=df_2d);
```
