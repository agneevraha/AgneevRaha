# Matplotlib

```python
# import libraries
import pandas as pd 
import matplotlib.pyplot as plt
```

## Load `mtcars` Dataset

```python
df = pd.read_csv('mtcars.csv')
```

## Check the Data

```python
df.head()
```

```python
df.shape
```

## Univariate Plots Using Matplotlib

```python
# boxplot
plt.boxplot(df['mpg'],patch_artist=True)
plt.title('Distribution of Miles Per Gallon')
plt.ylabel('miles per gallon');
```

```python
# histogram
plt.hist(df['mpg'],edgecolor='white')
plt.title('Histogram of Mpg')
plt.ylabel('Count of Observation')
plt.xlabel('bins of mpg values');
```

```python
# Lineplot
plt.plot(df['mpg']);
```

## Bivariate Plots using Matplotlib

```python
# scatterplot - weight vs mileage
plt.scatter(df['wt'],df['mpg'])
plt.title('scatterplot of weight vs miles per gallon')
plt.xlabel('weight of the vehicle')
plt.ylabel('miles per gallon');
```

```python
# scatterplot - displacement vs horsepower
plt.scatter(df['disp'],df['hp']);
```

```python
plt.scatter(df['carb'],df['mpg']);
```

```python
plt.scatter(df['disp'],df['mpg']);
```

```python
# practice scatter plots between other pairs of numeric features in the data
```

```python
df.corr().round(2)
```

## Subplots Using Matplotlib

```python
# plt.subplots
fig,axes = plt.subplots(ncols=2, nrows=1)
axes[0].boxplot(df['mpg'])
axes[1].hist(df['hp'])
plt.suptitle('Distribution of mpg and hp')
axes[0].title.set_text('mpg')
axes[1].title.set_text('hp')
plt.tight_layout();
```

```python
# plt.subplot
plt.subplot(121)
plt.boxplot(df['mpg'])
plt.subplot(122)
plt.hist(df['hp']);
```

# Seaborn Library

```python
import seaborn as sns
```

## Load tips Data inbuilt in Seaborn

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

## Check the head of the data

```python
df.head()
```

## Univariate Plots using Seaborn

```python
# boxplot
sns.boxplot(data=df,x='total_bill',color='gold');
```

```python
# histogram
sns.histplot(data=df, x='total_bill');
```

```python
# density plot
sns.kdeplot(data=df, x='total_bill',fill=True,color='green');
```

```python
df['day'].value_counts()
```

```python
df['day'].value_counts(normalize=True).round(2)*100
```

```python
df['sex'].value_counts()
```

```python
# countplot
sns.countplot(data=df, x='day')
plt.title('Bill Frequencies on days of the week');
```

```python
sns.countplot(data=df, x='day',order=['Sat','Sun','Thur','Fri'])
plt.title('Bill Frequencies on days of the week');
```

```python
sns.countplot(data=df, x='day',order=df['day'].value_counts().index)
plt.title('Bill Frequencies on days of the week');
```

```python
sns.countplot(data=df, x='sex');
```

```python
# countplot - ordered bars - practice for the other categorical features
```

```python
# sns.lineplot('tip',data=df);
```

```python
# lineplot
sns.lineplot(df['tip']);
```

```python
sns.lineplot(data=df, x=df.index, y='tip');
```

## Bivariate Plots using Seaborn

```python
# boxplot
sns.boxplot(data=df, x='sex',y='total_bill');
```

```python
# barplot
sns.barplot(data=df,x='sex',y='tip',errorbar=('ci', False))
plt.ylabel('Average Tip amount($)');
```

```python
# pointplot
sns.pointplot(data=df, x='sex',y='tip', errorbar=('ci',False));
```

```python
# density plots
sns.kdeplot(data=df, x='total_bill',hue='sex',fill=True);
```

```python
# pair plot
```

```python
# correlation matrix
corel = df.corr().round(2)
corel
```

```python
# heatmap
sns.heatmap(data=corel,
           annot=True,
           cmap='plasma',
           vmin=-1,
           vmax=1);
```
