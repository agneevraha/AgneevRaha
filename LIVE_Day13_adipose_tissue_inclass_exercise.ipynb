```python
# import usual libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# load the dataset wc_at.csv
df = pd.read_csv('Data Sets/wc_at.csv')
```

```python
# head of the dataset
df.head()
```

```python
# check the data types
df.info()
```

```python
# check the distribution of the data
fig, axes = plt.subplots(ncols=2, nrows=2,figsize=(10,5))
sns.boxplot(data=df, x='Waist',ax = axes[0,0])
sns.kdeplot(data=df, x='Waist',ax=axes[0,1])
sns.boxplot(data=df, x='AT', ax=axes[1,0])
sns.kdeplot(data=df, x='AT',ax=axes[1,1])
plt.tight_layout();
```

```python
# check the correlation between the variables
corel = df.corr().round(2)
corel
```

```python
# scatter plot of the data
sns.scatterplot(data=df, x='Waist', y='AT');
```

```python
# Separate the target variable from the independent variable(s)
X = df.drop('AT', axis=1)
y = df['AT']
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# train test split - 70% training , 30 % test
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=123)
```

```python
# create a linear regression model using sklearn & fit it on the data
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# get the prdictions of the model for train and test data observations
pred_train = lr.predict(X_train)
pred_test = lr.predict(X_test)
```

```python
# evaluate the model - r-squared metric
r2_train = lr.score(X_train, y_train)
r2_test = lr.score(X_test, y_test)
r2_train,r2_test
```
