```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from scipy.stats import probplot as pb
```

```python
df=pd.read_csv(r"C:\Users\hp\Desktop\ExcelR\Assignments Folder\1Basic Statistics Level_1\Q21 A Cars.csv")
```

```python
plt.hist(df)
```

```python
rvs =norm(loc=0, scale=1)
```

```python
df=rvs.rvs(size=300)
```

```python
sns.set()
sns.distplot(df)
```

```python
df=rvs.rvs(size=300)
pb(df, dist="norm", plot=plt)
plt.show()
```

```python
# import useful library
import numpy as np
from scipy.stats import shapiro
from numpy.random import randn
 
# Create data
df = randn(300)
 
# conduct the  Shapiro-Wilk Test
shapiro(df)
```

```python
#Using Null hypothesis, It shows that the data doesnt follow normal distribution.
```
